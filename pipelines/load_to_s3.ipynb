# LOAD Part 1: Upload Data to Amazon S3

```python
import boto3
import os

# Create an S3 client
s3 = boto3.resource('s3')
print(s3)
```

```python
# Check Existing Bucket
bucket_name = 'ds4a-team21-bucket'
folder_processed = 'processed'
folder_json_data = 'json_data'

for bucket in s3.buckets.all():
    print(bucket.name)
```

```python
# Upload processed data to S3
for file in os.listdir('../data/processed'):
    if file.endswith(".csv"):
        print(file)
        s3.meta.client.upload_file('../data/processed/'+file, bucket_name, folder_processed+'/'+file)
```

```python
# Upload json_data files to S3
for file in os.listdir("../data/raw/json_data"):
    if file.endswith(".json"):
        print(file)
        s3.meta.client.upload_file('../data/raw/json_data/'+file, bucket_name, folder_json_data+'/'+file)
```
